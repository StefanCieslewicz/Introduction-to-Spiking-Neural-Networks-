<h3> Introduction </h3>

Recent advancements in the in artificial intelligence can be largely attributed to neural networks (NNs), a machine learning model that recognizes patterns and makes predictions. Artificial neural networks consist of many layers, each consisting of "neurons", this type of a machine learning model is clearly inspired by biological neural networks, yet it remains fundamentally different from them. While research around classic NNs is not losing momentum, another type of neural networks is beginning to gain attention that is spiking neural networks (SNNs). They present a different approach to NN, they aim to emulate brain dynamics and energy efficiency of real neurons. 
As we progress further in the field of machine learning, we should shift our attention to unconventional computation methods and explore SNNs and neuromorphic computing - an approach to hardware and computation, strive to achieve more efficient and brain like artificial intelligence.
